```python
deploy1 = pd.read_pickle(os.path.join(DEPLOY1_FOLDER_PATH,"players.pkl"))
```

```python
deploy1.info()
```

```python
finalPlayers = pd.concat([players, deploy1],ignore_index=True)
```

```python
finalPlayers = finalPlayers.sort_values(["burnerParent","totalGames"], ascending = [False, False])
finalPlayers.reset_index(drop=True,inplace=True)
print(finalPlayers.info())
finalPlayers.head(10)
```

```python
#19481 - 1599 (those with no parent) = 17882
```

```python
finalPlayersWithParentBurner = finalPlayers[finalPlayers["burnerParent"] !="0x0000000000000000000000000000000000000000"]
print(finalPlayersWithParentBurner.info())
finalPlayersWithParentBurner.sort_values(by=["burnerParent", "totalGames"],ascending=[False,False],inplace=True)
finalPlayersWithParentBurner.head(10)
```

```python
# output FinalPlayersWithBurners
finalPlayersWithParentBurner.to_pickle("finalWithParents13102023.pkl")
```

```python
finalPlayersWithParentBurner.to_csv("finalWithParents13102023.csv",index=False)
```

```python
# get totals for parent burner
parentBurnerStats = finalPlayersWithParentBurner.groupby("burnerParent")[["totalWins","totalLosses", "totalGasEarned", "totalGasLost", "totalGames", "totalRevenue"
                                                                         ]].sum().reset_index()
print(parentBurnerStats.info())
parentBurnerStats.head(10)
```

```python
complete1Games = parentBurnerStats[parentBurnerStats["totalGames"]>0]
complete1Games.to_csv("complete1Games13102023.csv", index=False)
```

```python
complete3Games = parentBurnerStats[parentBurnerStats["totalGames"]>2]
complete3Games.to_csv("complete3Games13102023.csv", index=False)
```

```python
totalWins1Games = parentBurnerStats[parentBurnerStats["totalWins"]>0]
totalWins1Games.to_csv("totalWins1orMore13102023.csv", index=False)
```
